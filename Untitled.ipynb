```python
import numpy as np 
import pandas as pd 
import os
import tensorflow as tf
import PIL
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, GlobalAveragePooling2D, Dropout, Flatten
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
train_dir = 'data/train/'
test_dir = 'data/test/'
val_dir = 'data/valid/'
```

## Data preparataion and visualisation

```python
img_height,img_width=224,224
batch_size=16

train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
  train_dir,
  seed=123,
  shuffle=True,
  image_size=(img_height, img_width),
  batch_size=batch_size)

val_dataset = tf.keras.preprocessing.image_dataset_from_directory(
  val_dir,
  shuffle=False,
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
test_dataset = tf.keras.preprocessing.image_dataset_from_directory(
  test_dir,
  shuffle=False,
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
class_names = train_dataset.class_names
print(class_names)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15))
for images, labels in train_dataset.take(1):
    for i in range(10):
        ax = plt.subplot(5, 5, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")

plt.figure(figsize=(15, 15))
for images, labels in train_dataset.take(1):
    for i in range(10):
        ax = plt.subplot(5, 5, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

## Using Resnet50 as base model:

```python
from tensorflow.keras.applications import ResNet50
```

```python
base_model= tf.keras.applications.ResNet50(include_top=False,
                   input_shape=(224,224,3),
                   pooling='avg',classes=100,
                   weights='imagenet')
```

### Freezing the pre-trained layers of ResNet50

```python
for layer in base_model.layers:
        layer.trainable=False
```

```python
from tensorflow.keras.optimizers import Adam
```

## Building model: adding one Dense layer and one Softmax Layer

```python
model = Sequential()
```

```python
model.add(base_model)
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5)) # to prevent overfitting
model.add(Dense(525, activation='softmax'))
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001),
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.summary()
```

### Using GPU

```python
gpus = tf.config.list_physical_devices('GPU')
tf.config.set_visible_devices(gpus[0], 'GPU')
```

```python
train_dataset = train_dataset.cache().shuffle(300).prefetch(buffer_size=tf.data.AUTOTUNE)
val_dataset = val_dataset.cache().shuffle(300).prefetch(buffer_size=tf.data.AUTOTUNE)
```

## Training model

```python
history = model.fit(train_dataset,
                    epochs=15,
                    validation_data=val_dataset,
                    )
```

## Visualising learning curves for Model loss and Model accuracy

```python
plt.figure(figsize=(10,4))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend(['Train','Val'], loc= 'upper left')
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```

## Test dataset prediction accuracy

```python
test_dataset = test_dataset.cache().shuffle(300).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
result = model.evaluate(test_dataset)
print("Test loss, test accuracy:", result[1])
```

```python
predictions = model.predict(test_dataset)
score = tf.nn.softmax(predictions)
```
